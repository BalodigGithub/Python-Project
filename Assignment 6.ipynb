```python
Answer 1 :

Commonly used escape characters include:

\n: Represents a newline character. When this escape sequence is encountered within a string, it creates a new line, moving the
    
cursor to the beginning of the next line.

\t: Represents a tab character. It adds horizontal spacing, typically equivalent to a tab stop.

\: Represents a literal backslash character. If you need to include an actual backslash in a string, you escape it with another
backslash.

\r: Represents a carriage return character. It's used to move the cursor to the beginning of the current line.

# Using escape characters in a string
message = "Hello, \nWorld!"
print(message)

# Including double quotes inside a string
quote = "He said, \"Hello!\""
print(quote)

# Including a backslash
path = "C:\\Users\\Username\\Documents"
print(path)
The output of the above Python code would be:

makefile
Copy code
Hello,
World!
He said, "Hello!"
C:\Users\Username\Documents

    In summary, escape characters allow you to include special characters or control sequences within strings by prefixing them 
with a backslash, which tells the interpreter or compiler to treat them differently than their literal representation.
```

```python
Answer 2 :
The escape sequences \n and \t stand for the following:
    
1 ) \n: The \n escape sequence represents a newline character. When it appears in a string, it is used to create a new line or 
start a new line of text. It is also known as a "line feed" character. 
2 ) \t: The \t escape sequence represents a tab character. It is used to add horizontal spacing or indentation within a string.
When \t is encountered in a string, it inserts a tab character, which is typically equivalent to a tab stop's worth of 
horizontal space.
# Using \n and \t escape sequences
text = "This is the first line.\n\tThis is the second line with a tab.\nThis is the third line."
print(text)
```

```python
Answer 9 :
    
1 )Right-Justify: To right-justify a string, you can use the str.rjust(width, fillchar) method. This method pads the string 
with a specified character (default is a space) on the left side until it reaches the specified width. Here's an example:
text = "Hello"
justified_text = text.rjust(10, "-")
print(justified_text)
2 )Left-Justify: To left-justify a string, you can use the str.ljust(width, fillchar) method. This method pads the string with 
a specified character (default is a space) on the right side until it reaches the specified width. Here's an example
text = "Hello"
justified_text = text.ljust(10, "-")
print(justified_text)
```

```python
Answer 10 :
    
The best way to remove whitespace characters from the start or end of a string in Python is to use the str.strip() method. This
method removes leading and trailing whitespace characters, including spaces, tabs, and newline characters, by default. 
text = "   This is a string with leading and trailing spaces.   "
result = text.strip()
print(result)
The strip() method is efficient and easy to use, making it a widely accepted choice for removing leading and trailing whitespace
characters from a string. However, keep in mind that it removes all leading and trailing whitespace, which may or may not be 
what you want.
text = "   Leading spaces."
result = text.lstrip()
print(result)  # "Leading spaces."

text = "Trailing spaces.   "
result = text.rstrip()
print(result)  # "Trailing spaces."
    
```
