```python
Answer 6 :
    
from PyPDF2 import PdfFileReader, PdfFileWriter

# Open the PDF file in binary read mode
with open('input.pdf', 'rb') as pdf_file:
    pdf_reader = PdfFileReader(pdf_file)
    
    # Create a PdfFileWriter object to write the rotated page
    pdf_writer = PdfFileWriter()
    
    # Specify the page number you want to rotate (e.g., page 1)
    page_number = 0  # Page numbering is 0-based
    
    # Rotate the page 90 degrees clockwise
    page = pdf_reader.getPage(page_number)
    page.rotateClockwise(90)
    
    # Add the rotated page to the PdfFileWriter
    pdf_writer.addPage(page)
    
    # Write the rotated page to a new PDF file
    with open('output.pdf', 'wb') as output_pdf:
        pdf_writer.write(output_pdf)
```

```python
Answer 7 :
    
1) Paragraph Object:

A Paragraph object represents a single paragraph within a document. A paragraph is a block of text that typically ends with a
hard return (pressing the "Enter" or "Return" key).
It can contain multiple Run objects, making it a container for text and other elements such as formatting, styles, and
alignment.
Paragraphs are often used to organize and structure content within a document. You can apply paragraph-level formatting, like
indentation or spacing, to all the text within a paragraph.

2)Run Object:

A Run object represents a contiguous range of text within a paragraph. It is the smallest unit of text that can be styled or
formatted independently.
Runs can be used to apply specific character-level formatting, like font styles, text color, bold or italic formatting, or 
other styling options.
Multiple Run objects can exist within a single paragraph, each with its own formatting attributes.
```
